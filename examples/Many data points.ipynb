```python
from xyplot import Curve

# Our data
x, y = [99.98, 120.06, 140.76, 159.86, 179.86, 200, 220.40, 239.6, 260, 279.80, 300], [3.02, 3.25, 3.42, 3.58, 3.80, 3.95, 4.07, 4.25, 4.35, 4.52, 4.66]
# A simple curve object with data and degree of polynomial
curve = Curve(
    x, 
    [k*k for k in y], 
    1
)

# Set the x, y axis labels and Title
curve.set(
    xlabel = "Volts",
    ylabel = "Current^2",
    title = "Variation of Current^2 with Volts"
)

# Label our data and curve
curve.createPlot(
    colors=['#aaaaaa', '#000000'],
    plotLabel="r = 0.02m",
#     dataLabel="Label for our DATA",
)

x, y = [100.75, 119.93, 139.85, 160.59, 180.29, 200.30, 220.30, 239.70, 259.80, 279.80, 300.10], [1.87, 2.03, 2.20, 2.33, 2.45, 2.56, 2.66, 2.75, 2.82, 2.95, 3.04]

# Label our data and curve
curve.createPlot(
    data=(
        x,
        [k**2 for k in y],
        1
    ),
    colors=['#aaaaaa', '#333333'],
    plotLabel="r = 0.03m",
#     dataLabel="Label for our DATA",
)


x, y = [120.20, 139.92, 160.44, 180.20, 200.50, 220.50, 240.00, 259.70, 280.20, 299.50], [1.51, 1.64, 1.72, 1.80, 1.89, 1.96, 2.04, 2.11, 2.18, 2.24]

# Label our data and curve
curve.createPlot(
    data=(
        x,
        [k**2 for k in y],
        1
    ),
    colors=['#aaaaaa', '#666666'],
    plotLabel="r = 0.04m",
#     dataLabel="Label for our DATA",
)



x, y = [120.80, 139.20, 159.50, 180.10, 200.30, 220.30, 239.80, 260.30, 280.30, 299.80], [1.20, 1.31, 1.39, 1.46, 1.5, 1.58, 1.63, 1.69, 1.74, 1.79]

# Label our data and curve
curve.createPlot(
    data=(
        x,
        [k**2 for k in y],
        1
    ),
    colors=['#aaaaaa', '#999999'],
    plotLabel="r = 0.05m",
#     dataLabel="Label for our DATA",
)


# curve.save("ebym.png") # Save our graph in high quality
```
